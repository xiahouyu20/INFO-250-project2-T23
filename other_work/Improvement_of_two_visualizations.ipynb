# Visualization 2

### Original Visualization 2
What is it talking about?
- This visualization shows several economic indicators of China during the past 10 years, which is 2012-2021
- It adopts **Nightingale Rose Chart** to show the increase or decrease of each indicators in those years.
- The indicators shows are presented below:
  - Retail sales of consumer goods 
  - Sales of commercial housing 
  - Sales of residential housing
  - Sales of China's top 100 chain stores 
  - Sales of China's top 100 fast-moving consumer goods 
  - Total retail sales of alcoholic products from enterprises above designated size in China 
  - Total retail sales of snack foods in China
- For each indicator, only the data of 2012 and 2021 are provided, so we think the value of each indicator in each year is not what authors want convey to readers. **It is the trend that authors want to show**.

Why it is choosed?
- We know that **Nightingale Rose Chart** uses the radius of the arc to represent the size (amount) of the data. Since the relationship between radius and area is square, the **Nightingale rose chart** will exaggerate the scale of data, especially suitable for comparing values with similar sizes.
- This visualization is suitable for the objective, **however, a disaster appears when data of different orders of magnitude are put in one chart**
- For example, the first indicator *Retail sales of consumer goods* is twenty times of the last indicator *Total retail sales of snack foods in China*. **The huge different orders makes the last four indicators a 'truely' circle, it seems like those indicators haven't change during the past ten years, but remarkable increase are shown when you have a look on the data of 2012 and 2021, for *Total retail sales of snack foods in China*, it increased 2 times!**
