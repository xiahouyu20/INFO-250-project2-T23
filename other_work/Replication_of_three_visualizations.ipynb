```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
from math import pi
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from matplotlib.ticker import FuncFormatter
import matplotlib.gridspec as gridspec
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

fig, ax = plt.subplots(figsize=(8, 8))
ax = plt.subplot(projection='polar')
ax.set_theta_direction(-1)
data = [56, 59, 61, 62, 63, 65, 70, 75, 75]
startangle = -90
colors = ['#FEEAC2', '#C6E2B9', '#79CAB7', '#08B9B4', '#8DD7F7', '#6B85B8', '#1964AA', '#F0502B', '#0DAADC']
xs = [(i * pi * 2)/ 100 for i in data]
ys = [1, 2, 3, 4, 5, 6, 7, 8, 9]
left = (startangle * pi *2)/ 360   #this is to control where the bar starts

# plot bars and points at the end to make them round
for i, x in enumerate(xs):
    ax.barh(ys[i], x, left=left, height=0.9, color=colors[i], tick_label=i)

ax.text(-pi/2 - pi * 0.10, 9.2, 'Thaliand', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.11, 8.27, 'HKSAR', fontweight='bold', fontsize='x-large', color='#FF4500')
ax.text(-pi/2 - pi * 0.145, 7.6, 'Singapore', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.13, 6.3, 'Canada', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.25, 6.8, 'United Kindom', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.17, 4.5, 'France', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.31, 5, 'South Afirca', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.22, 2.3, 'UAE', fontweight='bold', fontsize='large')
ax.text(-pi/2 - pi * 0.38, 2.15, 'Japan', fontweight='bold', fontsize='large')

ax.text(pi/2 + pi * 0.08, 1.3, '56', fontweight='bold', fontsize='large')
ax.text(pi/2 + pi * 0.05, 2.2, '62', fontweight='bold', fontsize='large')
ax.text(pi/2 + pi * 0.036, 3.25, '63', fontweight='bold', fontsize='large')
ax.text(pi/2 + pi * 0.0275, 4.2, '64', fontweight='bold', fontsize='large')
ax.text(pi/2 + pi * 0.025, 5.2, '68', fontweight='bold', fontsize='large')
ax.text(pi/2 + pi * 0.02, 6.2, '69', fontweight='bold', color='white', fontsize='large')
ax.text(pi/2 + pi * 0.015, 7.2, '75', fontweight='bold', color='white', fontsize='large')
ax.text(pi/2 + pi * 0.02, 8.3, '80', fontweight='bold', color='white', fontsize='xx-large')
ax.text(pi/2 + pi * 0.011, 9.2, '80', fontweight='bold', color='white', fontsize='large')
    
plt.xticks([])
plt.yticks([])
ax.spines.clear()
plt.show()
```
